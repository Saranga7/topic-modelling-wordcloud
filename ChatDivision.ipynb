```python
import pandas as pd
```

```python
WholeChat = pd.read_csv("WholeChat.csv")
Dates = WholeChat['Date']
Dates.head()
```

```python
pure_date = []
for i in Dates:
    pure_date.append(i[0] + i[1])
pure_date[0:10]
```

```python
count = 0
last_date = pure_date[0]
c = 0
last_count = 0
for i in pure_date:
    if i != last_date:
        print(str(i) + ", " + str(count))
        last_date = i
        df = WholeChat[last_count:last_count+count]
        print(df['Date'])
        df.to_csv("ModChat" + str(c) + ".csv")
        c = c + 1
        last_count = last_count + count
        count = 0
    count = count + 1
    #print(i)
```
