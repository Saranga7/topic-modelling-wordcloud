```python
nlp=spacy.load('en_core_web_sm')
from spacy.lang.en import English
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
stops="good,sure,okay,great,hello,okay,ok,yeah,thanks,alright,please,lol,yes,cool,wow,hi,oh,pm"
temp=stops.split(',')
for i in temp:
    STOP_WORDS.add(i)
```

```python
df=[]

for i in range(9,39):
    x=pd.read_csv("Modulated Chat\ModChat"+str(i)+".csv")

    df.append(x)
    

```

```python
for i in range(0,30):
    df[i].drop(['index','Message_Raw','Message_Only_Text',"Date",'Hour','Day_of_Week'],axis=1,inplace=True)
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lem=WordNetLemmatizer()
corpus=[]
```

```python
for j in range(0,30):
    y=[]
    for i in range(0,len(df[j])):
        x=re.sub('[^a-zA-Z]',' ',df[j]['Message_Clean'][i])
        x=x.lower()
        x=x.split()

        x=[lem.lemmatize(word) for word in x if not word in stopwords.words('english')  and not word in STOP_WORDS]
        x=' '.join(x)
        y.append(x)
    corpus.append(y)
        
```

```python
string_list=[]

for i in corpus:
    strings=""
    for j in i:
        strings+=j
        strings+=' '
    string_list.append(strings)
```

```python
j=1
for i in string_list:
    wordcloud = WordCloud(background_color = "white", max_words = 10,contour_width = 3, contour_color = 'steelblue')    
    wordcloud.generate(i)
    wordcloud.to_image()
    wordcloud.to_file('cloud'+str(j)+'.png')
    j+=1
```
